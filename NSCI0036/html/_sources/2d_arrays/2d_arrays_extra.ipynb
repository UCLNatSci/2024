## Scalar Fields

A scalar field is a mapping which associates a number with points in space. Examples of scalar fields area atmospheric pressure (measured in millibars) of points in a geographical area, or  the temperature (in Kelvin) of points in a 3 dimensional solid body.

Suppose we have a 2-dimensional scalar field given by the following formula:

$$
\begin{align}
  f(x, y) = 
  \begin{cases}
    \exp(-y^2) & \text{if} -1.5 < x < 1.5 \\
    0 & \text{otherwise.}
  \end{cases}
\end{align}
$$

````python
import matplotlib.pyplot as plt
import numpy as np

def height(x, y):
    val = np.exp(-y**2)      
    if x < 1.5 and x > -1.5:
        result = val
    else:
        result = 0    
    return result
N = 30

z = np.zeros((N, N))

x_array = np.linspace(-2, 2, N)
y_array = np.linspace(-2, 2, N)

for j in range(N):
    for i in range(N):
        x = x_array[i]
        y = y_array[j]
        z[i,j] = height(x, y)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$f(x,y)$")


# Plot the surface
x1, y1 = np.meshgrid(x_array, y_array)
ax.plot_surface(x1, y1, z);
````
